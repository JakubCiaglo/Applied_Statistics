# Importy

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

___
# Zadanie 1

```python
MC = 1000
X = np.random.exponential(1/2, size=MC)
Y = np.random.exponential(1, size=MC)
Z = np.maximum(X, Y)
W = np.minimum(X, Y)

def dystrybuanta_z(z):
    return np.where(z < 0, 0, 1 - np.exp(-z) - np.exp(-2*z) + np.exp(-3*z))

def dystrybuanta_w(w):
    return np.where(w < 0, 0, 1 - np.exp(-3*w))

dane_x = np.linspace(0, 5, 1000)
teoretycznie_z = dystrybuanta_z(dane_x)
teoretycznie_w = dystrybuanta_w(dane_x)

plt.figure(figsize=(10, 6))
plt.plot(dane_x, teoretycznie_z, label='Teoretyczna', color='blue')
sns.ecdfplot(Z, label='Empiryczna', color='red')
plt.ylim(0, 1.1)
plt.title('Dystrybuanta Z')
plt.xlabel('Z')
plt.ylabel('F_Z(z)')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(dane_x, teoretycznie_w, label='Teoretyczna', color='blue')
sns.ecdfplot(W, label='Empiryczna', color='red')
plt.ylim(0, 1.1)
plt.title('Dystrybuanta W')
plt.xlabel('W')
plt.ylabel('F_W(w)')
plt.legend()
plt.grid(True)
plt.show()

def gestosc_z(z):
    return np.where(z < 0, 0, 2*np.exp(-2*z) + np.exp(-z) - 3*np.exp(-3*z))

def gestosc_w(w):
    return np.where(w < 0, 0, 3*np.exp(-3*w))

gestosc_teoretyczna_z = gestosc_z(dane_x)
gestosc_teoretyczna_w = gestosc_w(dane_x)

plt.figure(figsize=(10, 6))
plt.plot(dane_x, gestosc_teoretyczna_z, label='Teoretyczna', color='blue')
sns.kdeplot(Z, label='Empiryczna', color='red', fill=True, alpha=0.3)
plt.ylim(0, 1)
plt.title('Gęstość Z')
plt.xlabel('Z')
plt.ylabel('f_Z(z)')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(dane_x, gestosc_teoretyczna_w, label='Teoretyczna', color='blue')
sns.kdeplot(W, label='Empiryczna', color='red', fill=True, alpha=0.3)
plt.ylim(0, 3.5)
plt.title('Gęstość W')
plt.xlabel('W')
plt.ylabel('f_W(w)')
plt.legend()
plt.grid(True)
plt.show()
```

# Zadanie 2


```python
x = [4,10,7.5,9,11,2,5,1,3,7.5,6]
y = [5,11,6,8.5,10,2,3,1,4,7,8.5]
d = [x[i] - y[i] for i in range(11)]

def korelacja(d):
    n = len(d)
    suma = 0
    for i in d:
        suma += i**2
    return 1 - (6*suma)/(n*(n**2 - 1))
korelacja(d)
```
