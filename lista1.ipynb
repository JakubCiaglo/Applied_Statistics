```python
x = open('dane.txt')
iksy = x.read().split()

dane_x = []
dane_y = []

for i in iksy:
    i = float(i)
    dane_x.append(i)
    dane_y.append(np.exp(i))

plt.figure(figsize=(10, 6))
plt.hist(dane_x, bins=30, color='skyblue', edgecolor='black')
plt.title('Rozkład X', fontsize=14)
plt.xlabel('Wartości X', fontsize=12)
plt.ylabel('Liczność', fontsize=12)
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(dane_y, bins=30, color='salmon', edgecolor='black')
plt.title('Rozkład Y = exp(X)', fontsize=14)
plt.xlabel('Wartości Y', fontsize=12)
plt.ylabel('Liczność', fontsize=12)
plt.grid(True)
plt.show()
```

```python
dane = open('dane.txt')
xs = dane.read().split()

dane_x = []
for i in xs:
    i = float(i)
    dane_x.append(i)

dane_y = []
for i in xs:
    i = float(i)
    i = np.exp(i)
    dane_y.append(i)

x_range = np.linspace(-10, 30, 100)
X = norm.pdf(x_range, np.mean(dane_x), np.std(dane_x))

y_range = np.linspace(-10, 30, 100)
Y = lognorm.pdf(y_range, s=np.std(dane_x), scale=np.exp(np.mean(dane_x)))

plt.figure(figsize=(10, 6))
plt.plot(x_range, X, label='Rozkład normalny', color='blue')
plt.hist(dane_x, density=True, bins=30, alpha=0.6, color='orange', label='Dane X')
plt.title('Rozkład X', fontsize=14)
plt.xlabel('Wartości X', fontsize=12)
plt.ylabel('Prawdopodobieństwo', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(y_range, Y, label='Rozkład log-normalny', color='green')
plt.hist(dane_y, density=True, bins=40, alpha=0.6, color='red', label='Dane Y')
plt.title('Rozkład Y', fontsize=14)
plt.xlabel('Wartości Y', fontsize=12)
plt.ylabel('Prawdopodobieństwo', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
dane = open('dane.txt')
xs = dane.read().split()

dane_x = []
for i in xs:
    i = float(i)
    dane_x.append(i)

dane_y = []
for i in xs:
    i = float(i)
    i = np.exp(i)
    dane_y.append(i)

x_range = np.linspace(-10, 30, 100)
X = norm.pdf(x_range, np.mean(dane_x), np.std(dane_x))

y_range = np.linspace(-10, 30, 100)
Y = lognorm.pdf(y_range, s=np.std(dane_x), scale=np.exp(np.mean(dane_x)))


plt.figure(figsize=(10, 6))
plt.plot(x_range, X, label='Teoretycznie', color='blue')
sns.kdeplot(dane_x, label="Empirycznie", color='orange')
plt.title('Rozkład X', fontsize=14)
plt.xlabel('Wartości X', fontsize=12)
plt.ylabel('Prawdopodobieństwo', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(y_range, Y, label='Teoretycznie', color='green')
sns.kdeplot(dane_y, label="Empirycznie", color='red')
plt.title('Rozkład Y', fontsize=14)
plt.xlabel('Wartości Y', fontsize=12)
plt.ylabel('Prawdopodobieństwo', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

# Zadanie 3

```python
def dystrubuanta(x):
    return 1 - np.power((lambda_ / (lambda_ + x)), alfa_)

n = 1000
lambda_ = 1
alfa_ = 3
wartosci = np.random.uniform(0, 1, n)
pareto = lambda_ * np.power(wartosci, -1 / alfa_) - lambda_

dane = pareto
posortowane = np.sort(dane)
empirycznie = np.arange(1, len(posortowane) + 1) / len(posortowane)
teoretycznie = dystrubuanta(posortowane)

plt.figure(figsize=(10, 6))
plt.plot(posortowane, empirycznie, label='Empirycznie', color='blue')
plt.plot(posortowane, teoretycznie, label='Teoretycznie', color='red', linestyle='--')
plt.title('Porównanie dystrybuanty empirycznej i teoretycznej', fontsize=14)
plt.xlabel('Wartości rozkładu Pareto', fontsize=12)
plt.ylabel('Dystrybuanta', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

# Zadanie 4

```python
n = 1000
lambda_ = 1
alfa_ = 3

wartosci = np.random.uniform(0, 1, n)
pareto = lambda_ * np.power(wartosci, -1 / alfa_) - lambda_

def teoretyczna_gestosc(x):
    return (alfa_ * lambda_**alfa_) / ((x + lambda_)**(alfa_ + 1))

wartosci_x = np.linspace(min(pareto), max(pareto), 1000)
teoretycznie = teoretyczna_gestosc(wartosci_x)

plt.figure(figsize=(10, 6))
sns.kdeplot(pareto, label='Gęstość empiryczna', color='blue')
plt.plot(wartosci_x, teoretycznie, 'r--', label='Gęstość teoretyczna')
plt.xlabel('Wartości')
plt.ylabel('Gęstość')
plt.title('Porównanie gęstości empirycznej i teoretycznej')
plt.legend()
plt.grid(True)
plt.show()
```

# Zadanie 5

```python
n = 1000
lambda_ = 1
alfa_ = 3
tau = 2
wartosci = np.random.uniform(0,1,n)
pareto = lambda_ * np.power(wartosci,-1/alfa_) - lambda_
Z = pareto
W = Z**(1/tau)
```

```python
def gestosc_teoretyczna(w):
    return (alfa_ * tau * lambda_**alfa_ * w**(tau - 1)) / ((lambda_ + w**tau)**(alfa_ + 1))

W_sorted = np.sort(W)
w_values = np.linspace(min(W_sorted), max(W_sorted), 1000)

plt.figure(figsize=(10, 6))
sns.kdeplot(W, label='Gęstość empiryczna', color='blue')
plt.plot(w_values, gestosc_teoretyczna(w_values), label='Gęstość teoretyczna', linestyle='--', color='red')
plt.xlabel('Wartości')
plt.ylabel('Gęstość')
plt.title('Porównanie gęstości empirycznej i teoretycznej dla zmiennej losowej W')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def dystrubuanta_teoretyczna(x):
    return 1 - np.power((lambda_ / (lambda_ + np.power(x, tau))), alfa_)

sorted_W = np.sort(W)
empirycznie = np.arange(1, len(sorted_W) + 1) / len(sorted_W)
w_values = np.linspace(min(sorted_W), max(sorted_W), 1000)

plt.figure(figsize=(10, 6))
plt.plot(w_values, dystrubuanta_teoretyczna(w_values), label='Dystrybuanta teoretyczna', linestyle='--', color='red')
sns.lineplot(x=sorted_W, y=empirycznie, drawstyle='steps-post', label='Dystrybuanta empiryczna', color='blue')
plt.xlabel('Wartości')
plt.ylabel('Dystrybuanta')
plt.title('Porównanie dystrybuanty empirycznej i teoretycznej dla zmiennej losowej W')
plt.legend()
plt.grid(True)
plt.show()
```
