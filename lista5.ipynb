# Importy

```python
import numpy as np
import matplotlib.pyplot as plt
```

___
# Zadanie 1 

```python
n = 1000
N = np.arange(1000, 11000, 1000)
alfa = 1
beta = 2
lista_alfy = []
lista_bety = []

for i in N:
    lista_est1 = []
    lista_est2 = []
    for _ in range(n):
        X = np.random.gamma(scale=alfa, shape=beta, size=i)
        estymator_beta = (np.mean(X)**2) / ((1/len(X)) * sum(X**2) - (np.mean(X)**2))
        estymator_alfa = np.mean(X) / ((1/len(X)) * sum(X**2) - (np.mean(X)**2))
        lista_est1.append(estymator_alfa)
        lista_est2.append(estymator_beta)
    lista_alfy.append(lista_est1)
    lista_bety.append(lista_est2)

plt.figure(figsize=(10, 6))
boxprops = dict(facecolor='lightblue', color='blue')
flierprops = dict(marker='o', markerfacecolor='red', markersize=5, linestyle='none')
plt.boxplot(lista_alfy, patch_artist=True, boxprops=boxprops, flierprops=flierprops)
plt.axhline(y=alfa, linestyle='--', color='black')
plt.title('Estymator Alfa')
plt.xticks(np.arange(1, 11), ['1000', '2000', '3000', '4000', '5000', '6000', '7000', '8000', '9000', '10000'])
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
boxprops = dict(facecolor='lightgreen', color='green')
plt.boxplot(lista_bety, patch_artist=True, boxprops=boxprops, flierprops=flierprops)
plt.axhline(y=beta, linestyle='--', color='black')
plt.title('Estymator Beta')
plt.xticks(np.arange(1, 11), ['1000', '2000', '3000', '4000', '5000', '6000', '7000', '8000', '9000', '10000'])
plt.grid(True)
plt.show()
```

# Zadanie 2

```python
MC = 1000
n = 1000
lamb_ = 1
teta = 2
estymatory_wiarygodnosci = []
estymatory_momentu = []
for i in range(MC):
    X = np.random.exponential(scale = lamb_,size = n)
    Y = X+teta
    estymator_wiarygodnosc = min(Y)
    estymatory_wiarygodnosci.append(estymator_wiarygodnosc)

    estymator_moment = (sum(Y)/len(Y)) -1
    estymatory_momentu.append(estymator_moment)

plt.figure(figsize=(10,6))
boxprops = dict(facecolor='lightblue', color='blue')
flierprops = dict(marker='o', markerfacecolor='red', markersize=5, linestyle='none')
plt.boxplot([estymatory_wiarygodnosci, estymatory_momentu], patch_artist=True, boxprops=boxprops, flierprops=flierprops)
plt.xticks([1, 2], ['Estymator Najwikszej Wiarygodnoci', 'Estymator Momentu'])
plt.title('Estymatory')
plt.grid(True)
plt.show()
```

# Zadanie 3

```python
MC = 1000
n = 1000
a = 2

estymatory_wiarygodnosci = []
estymatory_momentu = []
for i in range(MC):

    U = np.random.uniform(0,1,size = n)
    X = np.power(U,1/(a+1))
    estymator_wiarygodnosc = -((len(X)/sum(np.log(X))) + 1)
    estymatory_wiarygodnosci.append(estymator_wiarygodnosc)

    estymator_moment = (1 - (2/len(X))*sum(X))/ ((sum(X)/len(X)) - 1)
    estymatory_momentu.append(estymator_moment)

plt.figure(figsize=(10,6))
boxprops = dict(facecolor='lightblue', color='blue')
flierprops = dict(marker='o', markerfacecolor='red', markersize=5, linestyle='none')
plt.boxplot([estymatory_wiarygodnosci, estymatory_momentu], patch_artist=True, boxprops=boxprops, flierprops=flierprops)
plt.xticks([1, 2], ['Estymator Najwikszej Wiarygodnoci', 'Estymator Momentu'])
plt.title('Estymatory')
plt.grid(True)
plt.show()
```
